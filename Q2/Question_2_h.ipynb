```python
import matplotlib.pyplot as plt
from scipy import signal
import numpy as np
```

```python
ecg_signal = np.load("/content/patient_ecg.npy")
```

```python
nperseg = 600
```

```python
frequencies, times, Sxx = signal.spectrogram(ecg_signal, fs=250, nperseg=nperseg)
```

```python
plt.figure(figsize=(10, 5))
plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx + 1e-10), shading='gouraud', cmap='jet')
plt.title('Spectrogram')
plt.ylabel('Frequency (Hz)')
plt.xlabel('Time (s)')
plt.colorbar(label='Intensity (dB)')
plt.tight_layout()
plt.show()
```
