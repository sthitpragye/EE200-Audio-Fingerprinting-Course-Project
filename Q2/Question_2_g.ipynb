```python
import numpy as np
```

```python
patient_ecg = np.load("/content/patient_ecg.npy") # Complete ECG Signal
template = np.load("/content/template.npy") # A single healthy heartbeat
```

```python
def findonset(ecgsignal, template, threshold):
  count = 0 # Used to iterate over ECG signal in batches of 200
  curr_beat = 0 # Index of the current beat assuming 0 indexing.
  onset_beat = -1 # Used to record the index of the first Arrhythmia heartbeat
  # ans = [] # Can be uncommented to store all the correlation scores as a function of heartbeat index
  while(count<=len(ecgsignal)-len(template)):
    p = np.sum(template*ecgsignal[count:count+len(template)])/(np.sqrt(np.sum(np.square(template)))*np.sqrt(np.sum(np.square(ecgsignal[count:count+len(template)]))))
    # ans.append(p) # Can be uncommented to store all the correlation scores as a function of heartbeat index
    if(p<threshold and onset_beat==-1):
      onset_beat = curr_beat
      return onset_beat
    curr_beat+=1
    count+=len(template)
  return onset_beat
```

```python
onset_beat = findonset(patient_ecg, template, 0.5)
print(f"Index of heartbeat with Arrhythmia onset: {onset_beat}")
```
